```python
# Understanding the relationship between the Medical doctors(number), Generalist medical practitioners (number),
# Specialist medical practitioners (number), and Medical doctors not further defined (number)  

nan_records = df[df["Medical doctors (number)"].isna()]

# Show the result
print("\nRecords where 'Medical doctors not further defined (number)' is NaN:")
print(nan_records)
```

```python
# Replacing Missing values

df["Medical doctors (number)"] = df["Medical doctors (number)"].fillna(
    df["Generalist medical practitioners (number)"].fillna(0) +
    df["Specialist medical practitioners (number)"].fillna(0) +
    df["Medical doctors not further defined (number)"].fillna(0))

df["Generalist medical practitioners (number)"] = df["Generalist medical practitioners (number)"].fillna(
    df["Medical doctors (number)"] -
    df["Specialist medical practitioners (number)"].fillna(0) -
    df["Medical doctors not further defined (number)"].fillna(0))

df["Specialist medical practitioners (number)"] = df["Specialist medical practitioners (number)"].fillna(
    df["Medical doctors (number)"] -
    df["Generalist medical practitioners (number)"].fillna(0) -
    df["Medical doctors not further defined (number)"].fillna(0))

df["Medical doctors not further defined (number)"] = df["Medical doctors not further defined (number)"].fillna(
    df["Medical doctors (number)"] -
    df["Generalist medical practitioners (number)"].fillna(0) -
    df["Specialist medical practitioners (number)"].fillna(0))
```

```python
# Checking further missing values

df_nulls = df.isnull().sum()
print(f"Displaying all sum of null values: ")
print(f"{df_nulls}")
```

```python
# finding all the unique countries and unique year

all_countries = df['Countries, territories and areas'].unique().tolist()
print(f"All the countries included in the dataset: ", df['Countries, territories and areas'].nunique())
print(all_countries)
```

```python
all_year = sorted(df['Year'].unique().tolist())
print(f"All the year included in the dataset: ", df['Year'].nunique())
print(all_year)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# (optional) Set Seaborn style for nicer plots
sns.set(style="whitegrid")
```

```python
## Visualize Missing Values.
plt.figure(figsize=(10,8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values in Dataset')
plt.show()
```

Bright lines = missing data,
Dark lines = data is available

```python
# Calculate missing percentage
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage = missing_percentage[missing_percentage > 0].sort_values(ascending=False)

# Plot missing percentages
plt.figure(figsize=(10,5))
missing_percentage.plot(kind='barh', color='skyblue')
plt.title('Percentage of Missing Values by Column')
plt.xlabel('Percentage')
plt.ylabel('Columns')
plt.show()
```

```python
# List of numerical columns
numerical_cols = [
    'Medical doctors (per 10 000 population)',
    'Medical doctors (number)',
    'Generalist medical practitioners (number)',
    'Specialist medical practitioners (number)',
    'Medical doctors not further defined (number)'
]

# Plot histograms
plt.figure(figsize=(18, 12))
for idx, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 2, idx)
    sns.histplot(df[col].dropna(), kde=True, bins=30, color='teal')
    plt.title(f'Distribution of {col}', fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot boxplots for each numerical column
plt.figure(figsize=(18, 12))
for idx, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 2, idx)
    sns.boxplot(x=df[col], color='coral')
    plt.title(f'Boxplot of {col}', fontsize=12)
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
corr = df[numerical_cols].corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap among Doctor-related Metrics')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = "Doctors_by_countries.csv"
df = pd.read_csv(file_path)

# Ensure year is treated as integer
df['Year'] = df['Year'].astype(int)

# 1. Trend over time: Doctors per 10,000 people for a specific country
country_sample = "Nepal"
df_country = df[df['Countries, territories and areas'] == country_sample]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_country, x='Year', y='Medical doctors (per 10 000 population)', marker='o')
plt.title(f"Trend of Medical Doctors per 10,000 People in {country_sample}")
plt.xlabel("Year")
plt.ylabel("Doctors per 10,000 People")
plt.gca().invert_xaxis()  # To show most recent year on the left
plt.grid(True)
plt.show()
```

Trend over time for Nepal — you can see how doctors per 10,000 people changed year by year.

```python
# 2. Comparison by country: Number of doctors in the most recent year
most_recent_year = df['Year'].max()
df_recent = df[df['Year'] == most_recent_year]

# Top 15 countries for better visualization
df_recent_sorted = df_recent.sort_values(by='Medical doctors (number)', ascending=False).head(15)

plt.figure(figsize=(12, 8))
sns.barplot(data=df_recent_sorted, x='Medical doctors (number)', y='Countries, territories and areas', palette='viridis')
plt.title(f"Top 15 Countries by Number of Doctors in {most_recent_year}")
plt.xlabel("Number of Medical Doctors")
plt.ylabel("Country")
plt.grid(axis='x')
plt.show()
```

First chart: Trend over time for Nepal — you can see how doctors per 10,000 people changed year by year.

Second chart: Top 15 countries with the highest number of doctors in the most recent year.

```python
# 3. Scatter plot: Number of doctors vs doctors per 10,000 population
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_recent,
    x='Medical doctors (number)',
    y='Medical doctors (per 10 000 population)',
    hue='Countries, territories and areas',
    legend=False,
    s=100
)
plt.title(f"Doctors Number vs. Doctors per 10,000 Population ({most_recent_year})")
plt.xlabel("Number of Medical Doctors")
plt.ylabel("Doctors per 10,000 Population")
plt.xscale('log')  # Log scale because numbers vary greatly
plt.grid(True)
plt.show()
```

Scatter plot showing the relationship between the total number of doctors and doctors per 10,000 population — using a log scale for clarity.
